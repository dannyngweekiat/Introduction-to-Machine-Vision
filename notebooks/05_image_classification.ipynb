## 1. Load a picture
This lesson downloads MobileNetV2 weights the first time it runs. In Kaggle, ensure **Internet** is enabled.


```python
# BEGINNER NOTE: Import numpy so this notebook can use its ready-made tools.
import numpy as np
# BEGINNER NOTE: Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# BEGINNER NOTE: Import the specific tools this lesson needs from skimage.
from skimage import data
# BEGINNER NOTE: Import the specific tools this lesson needs from tensorflow.keras.applications.mobilenet_v2.
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions

# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
image = data.chelsea()
# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
plt.figure(figsize=(6, 4))
# BEGINNER NOTE: Display this image or result in the current plotting area.
plt.imshow(image); plt.axis('off');
```

## 2. Ask a pre-trained model
The model has learned from many labelled pictures. It returns several possible labels, each with a confidence score.


```python
# BEGINNER NOTE: Load a pre-trained image-classification AI model.
model = MobileNetV2(weights='imagenet')
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
resized = np.array(plt.imread('/kaggle/input/nonexistent.png')) if False else image  # keeps this notebook self-contained

# MobileNet expects a 224 x 224 image.
# BEGINNER NOTE: Import cv2 so this notebook can use its ready-made tools.
import cv2
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
model_input = cv2.resize(image, (224, 224)).astype('float32')
# BEGINNER NOTE: Prepare the image numbers in the format this AI model expects.
model_input = preprocess_input(model_input[None, ...])
# BEGINNER NOTE: Ask the trained AI model to make predictions for these examples.
predictions = decode_predictions(model.predict(model_input, verbose=0), top=3)[0]

# BEGINNER NOTE: Repeat the indented instructions once for every item in this group.
for _, label, confidence in predictions:
    # BEGINNER NOTE: Print a helpful result so you can inspect what the program found.
    print(f'{label:20} {confidence:.1%}')
```

## 3. Confidence is not certainty
A high score means “this looks similar to the model’s training examples.” It does not prove the answer is correct.


```python
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
labels = [label.replace('_', ' ') for _, label, _ in predictions]
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
scores = [score for _, _, score in predictions]
# BEGINNER NOTE: Draw a bar chart to compare the model's prediction scores.
plt.barh(labels[::-1], scores[::-1], color='#3D8DFF')
# BEGINNER NOTE: Set the horizontal chart scale from 0 to 1 for confidence scores.
plt.xlim(0, 1)
# BEGINNER NOTE: Label the horizontal axis so the chart is easier to understand.
plt.xlabel('Model confidence')
# BEGINNER NOTE: Add a title so the picture or result is easy to identify.
plt.title('Top three suggestions');
```
