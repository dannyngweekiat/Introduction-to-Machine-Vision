## 1. Load a picture
This lesson downloads MobileNetV2 weights the first time it runs. In Kaggle, ensure **Internet** is enabled.


```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions

image = data.chelsea()
plt.figure(figsize=(6, 4))
plt.imshow(image); plt.axis('off');
```

## 2. Ask a pre-trained model
The model has learned from many labelled pictures. It returns several possible labels, each with a confidence score.


```python
model = MobileNetV2(weights='imagenet')
resized = np.array(plt.imread('/kaggle/input/nonexistent.png')) if False else image  # keeps this notebook self-contained

# MobileNet expects a 224 x 224 image.
import cv2
model_input = cv2.resize(image, (224, 224)).astype('float32')
model_input = preprocess_input(model_input[None, ...])
predictions = decode_predictions(model.predict(model_input, verbose=0), top=3)[0]

for _, label, confidence in predictions:
    print(f'{label:20} {confidence:.1%}')
```

## 3. Confidence is not certainty
A high score means “this looks similar to the model’s training examples.” It does not prove the answer is correct.


```python
labels = [label.replace('_', ' ') for _, label, _ in predictions]
scores = [score for _, _, score in predictions]
plt.barh(labels[::-1], scores[::-1], color='#3D8DFF')
plt.xlim(0, 1)
plt.xlabel('Model confidence')
plt.title('Top three suggestions');
```
