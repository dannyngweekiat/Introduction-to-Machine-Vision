## 1. Load a picture
This lesson downloads MobileNetV2 weights the first time it runs. In Google Colab, keep the runtime connected to the internet for the first run.


```python
# Import numpy so this notebook can use its ready-made tools.
import numpy as np
# Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# Import the specific tools this lesson needs from skimage.
from skimage import data
# Import the specific tools this lesson needs from tensorflow.keras.applications.mobilenet_v2.
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions

# Store this result in a named variable so later lines can use it.
image = data.chelsea()
# Create a plotting area where the image results will be displayed.
plt.figure(figsize=(6, 4))
# Display this image or result in the current plotting area.
plt.imshow(image); plt.axis('off');
```

## 2. Ask a pre-trained model
The model has learned from many labelled pictures. It returns several possible labels, each with a confidence score.


```python
# Load a pre-trained image-classification AI model.
model = MobileNetV2(weights='imagenet')
# Store this result in a named variable so later lines can use it.

# MobileNet expects a 224 x 224 image.
# Import cv2 so this notebook can use its ready-made tools.
import cv2
# Store this result in a named variable so later lines can use it.
model_input = cv2.resize(image, (224, 224)).astype('float32')
# Prepare the image numbers in the format this AI model expects.
model_input = preprocess_input(model_input[None, ...])
# Ask the trained AI model to make predictions for these examples.
predictions = decode_predictions(model.predict(model_input, verbose=0), top=3)[0]

# Repeat the indented instructions once for every item in this group.
for _, label, confidence in predictions:
    # Print a helpful result so you can inspect what the program found.
    print(f'{label:20} {confidence:.1%}')
```

## 3. Confidence is not certainty
A high score means “this looks similar to the model’s training examples.” It does not prove the answer is correct.


```python
# Store this result in a named variable so later lines can use it.
labels = [label.replace('_', ' ') for _, label, _ in predictions]
# Store this result in a named variable so later lines can use it.
scores = [score for _, _, score in predictions]
# Draw a bar chart to compare the model's prediction scores.
plt.barh(labels[::-1], scores[::-1], color='#3D8DFF')
# Set the horizontal chart scale from 0 to 1 for confidence scores.
plt.xlim(0, 1)
# Label the horizontal axis so the chart is easier to understand.
plt.xlabel('Model confidence')
# Add a title so the picture or result is easy to identify.
plt.title('Top three suggestions');
```
