## 1. Classification versus detection
Classification answers “what might be in this image?” Detection also answers “where is it?”

This notebook downloads a small YOLO model on its first run. In Kaggle, ensure **Internet** is enabled.


```python
!pip -q install ultralytics

import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from ultralytics import YOLO

image = data.astronaut()
plt.figure(figsize=(6, 6))
plt.imshow(image); plt.axis('off');
```

## 2. Run the detector
Boxes are predictions. The number beside a box is a confidence score, not a guarantee.


```python
model = YOLO('yolov8n.pt')
results = model(image, conf=0.35, verbose=False)

annotated_bgr = results[0].plot()
annotated_rgb = annotated_bgr[:, :, ::-1]
plt.figure(figsize=(7, 7))
plt.imshow(annotated_rgb)
plt.axis('off');
```

## 3. Count selected objects
We choose a minimum confidence so very uncertain guesses do not count.


```python
names = results[0].names
classes = results[0].boxes.cls.cpu().numpy().astype(int)
confidences = results[0].boxes.conf.cpu().numpy()

for class_id, confidence in zip(classes, confidences):
    print(f'{names[class_id]}: {confidence:.1%}')

people = sum(names[class_id] == 'person' and confidence >= 0.35 for class_id, confidence in zip(classes, confidences))
print('People counted:', people)
```
