## 1. Classification versus detection
Classification answers “what might be in this image?” Detection also answers “where is it?”

This notebook downloads a small YOLO model on its first run. In Kaggle, ensure **Internet** is enabled.


```python
# Install the extra AI library needed for this activity.
!pip -q install ultralytics

# Import numpy so this notebook can use its ready-made tools.
import numpy as np
# Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# Import the specific tools this lesson needs from skimage.
from skimage import data
# Import the specific tools this lesson needs from ultralytics.
from ultralytics import YOLO

# Store this result in a named variable so later lines can use it.
image = data.astronaut()
# Create a plotting area where the image results will be displayed.
plt.figure(figsize=(6, 6))
# Display this image or result in the current plotting area.
plt.imshow(image); plt.axis('off');
```

## 2. Run the detector
Boxes are predictions. The number beside a box is a confidence score, not a guarantee.


```python
# Load or run an AI model that can find objects in an image.
model = YOLO('yolov8n.pt')
# Store this result in a named variable so later lines can use it.
results = model(image, conf=0.35, verbose=False)

# Create a copy of the picture with the AI's detection boxes drawn on it.
annotated_bgr = results[0].plot()
# Store this result in a named variable so later lines can use it.
annotated_rgb = annotated_bgr[:, :, ::-1]
# Create a plotting area where the image results will be displayed.
plt.figure(figsize=(7, 7))
# Display this image or result in the current plotting area.
plt.imshow(annotated_rgb)
# Hide the graph axes because they are not useful for this picture.
plt.axis('off');
```

## 3. Count selected objects
We choose a minimum confidence so very uncertain guesses do not count.


```python
# Store this result in a named variable so later lines can use it.
names = results[0].names
# Store this result in a named variable so later lines can use it.
classes = results[0].boxes.cls.cpu().numpy().astype(int)
# Store this result in a named variable so later lines can use it.
confidences = results[0].boxes.conf.cpu().numpy()

# Repeat the indented instructions once for every item in this group.
for class_id, confidence in zip(classes, confidences):
    # Print a helpful result so you can inspect what the program found.
    print(f'{names[class_id]}: {confidence:.1%}')

# Store this result in a named variable so later lines can use it.
people = sum(names[class_id] == 'person' and confidence >= 0.35 for class_id, confidence in zip(classes, confidences))
# Print a helpful result so you can inspect what the program found.
print('People counted:', people)
```
