# Lesson 2: Filters and Colour Masks

## Your mission
Make details clearer, find edges, and highlight one colour in an image.

**Key words:** blur · edge · mask

Run each code cell in order. Cells marked **Try it yourself** are yours to change.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

image = data.chelsea()  # RGB cat photograph
plt.figure(figsize=(6, 4))
plt.imshow(image)
plt.axis('off');
```

## 1. Blur and edges
A blur smooths tiny changes. Edge detection looks for sudden changes in brightness.


```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blurred = cv2.GaussianBlur(gray, (11, 11), 0)
edges = cv2.Canny(gray, 70, 160)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, picture, title in zip(axes, [gray, blurred, edges], ['Grayscale', 'Blurred', 'Edges']):
    ax.imshow(picture, cmap='gray')
    ax.set_title(title)
    ax.axis('off')
plt.show()
```

## 2. Make a colour mask
HSV represents colour in a way that makes it easier to choose a colour range.


```python
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# These numbers select orange-brown fur. They are not magic: we test and improve them.
lower = np.array([5, 70, 45])
upper = np.array([25, 255, 255])
mask = cv2.inRange(hsv, lower, upper)
highlighted = cv2.bitwise_and(image, image, mask=mask)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, picture, title in zip(axes, [image, mask, highlighted], ['Original', 'Colour mask', 'Highlighted pixels']):
    ax.imshow(picture, cmap='gray' if picture.ndim == 2 else None)
    ax.set_title(title)
    ax.axis('off')
plt.show()
print('Highlighted pixels:', np.count_nonzero(mask))
```
