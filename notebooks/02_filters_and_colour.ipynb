# Lesson 2: Filters and Colour Masks

## Your mission
Make details clearer, find edges, and highlight one colour in an image.

**Key words:** blur · edge · mask

Run each code cell in order. Cells marked **Try it yourself** are yours to change.


```python
# Import cv2 so this notebook can use its ready-made tools.
import cv2
# Import numpy so this notebook can use its ready-made tools.
import numpy as np
# Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# Import the specific tools this lesson needs from skimage.
from skimage import data

# Store this result in a named variable so later lines can use it.
image = data.chelsea()  # RGB cat photograph
# Create a plotting area where the image results will be displayed.
plt.figure(figsize=(6, 4))
# Display this image or result in the current plotting area.
plt.imshow(image)
# Hide the graph axes because they are not useful for this picture.
plt.axis('off');
```

## 1. Blur and edges
A blur smooths tiny changes. Edge detection looks for sudden changes in brightness.


```python
# Convert the image into a colour format that suits the next task.
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
# Blur nearby pixels together to reduce small details and noise.
blurred = cv2.GaussianBlur(gray, (11, 11), 0)
# Find strong changes in brightness, which often form edges.
edges = cv2.Canny(gray, 70, 160)

# Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# Repeat the indented instructions once for every item in this group.
for ax, picture, title in zip(axes, [gray, blurred, edges], ['Grayscale', 'Blurred', 'Edges']):
    # Display this image or result in the current plotting area.
    ax.imshow(picture, cmap='gray')
    # Add a title so the picture or result is easy to identify.
    ax.set_title(title)
    # Hide the graph axes because they are not useful for this picture.
    ax.axis('off')
# Show all of the plots you have prepared.
plt.show()
```

## 2. Make a colour mask
HSV represents colour in a way that makes it easier to choose a colour range.


```python
# Convert the image into a colour format that suits the next task.
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# These numbers select orange-brown fur. They are not magic: we test and improve them.
# Store this result in a named variable so later lines can use it.
lower = np.array([5, 70, 45])
# Store this result in a named variable so later lines can use it.
upper = np.array([25, 255, 255])
# Make a mask that keeps only pixels inside the chosen colour range.
mask = cv2.inRange(hsv, lower, upper)
# Use the mask to keep the selected pixels and hide the others.
highlighted = cv2.bitwise_and(image, image, mask=mask)

# Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# Repeat the indented instructions once for every item in this group.
for ax, picture, title in zip(axes, [image, mask, highlighted], ['Original', 'Colour mask', 'Highlighted pixels']):
    # Display this image or result in the current plotting area.
    ax.imshow(picture, cmap='gray' if picture.ndim == 2 else None)
    # Add a title so the picture or result is easy to identify.
    ax.set_title(title)
    # Hide the graph axes because they are not useful for this picture.
    ax.axis('off')
# Show all of the plots you have prepared.
plt.show()
# Print a helpful result so you can inspect what the program found.
print('Highlighted pixels:', np.count_nonzero(mask))
```
