## 1. Load a picture
An image is a grid of tiny colour squares called **pixels**. A colour pixel usually stores three numbers: red, green, and blue.


```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

image = data.astronaut()  # an RGB image included with scikit-image
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis('off')
plt.show()

print('Image shape (height, width, colour channels):', image.shape)
```

## 2. Look closely
The first two numbers tell us the image size. The final **3** means red, green, and blue channels.


```python
y, x = 180, 250
print(f'Pixel at (x={x}, y={y}) has RGB values:', image[y, x])

plt.figure(figsize=(5, 5))
plt.imshow(image)
plt.scatter([x], [y], color='yellow', s=90)
plt.title('One pixel, marked in yellow')
plt.axis('off');
```

## 3. Make simple edits


```python
gray = np.mean(image, axis=2)
crop = image[80:360, 130:410]
small = image[::3, ::3]
flipped = image[:, ::-1]

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, picture, title in zip(axes, [gray, crop, small, flipped], ['Grayscale', 'Crop', 'Smaller', 'Flipped']):
    ax.imshow(picture, cmap='gray' if picture.ndim == 2 else None)
    ax.set_title(title)
    ax.axis('off')
plt.show()
```
