## 1. Load a picture
An image is a grid of tiny colour squares called **pixels**. A colour pixel usually stores three numbers: red, green, and blue.


```python
# Import numpy so this notebook can use its ready-made tools.
import numpy as np
# Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# Import the specific tools this lesson needs from skimage.
from skimage import data

# Store this result in a named variable so later lines can use it.
image = data.astronaut()  # an RGB image included with scikit-image
# Create a plotting area where the image results will be displayed.
plt.figure(figsize=(6, 6))
# Display this image or result in the current plotting area.
plt.imshow(image)
# Hide the graph axes because they are not useful for this picture.
plt.axis('off')
# Show all of the plots you have prepared.
plt.show()

# Print a helpful result so you can inspect what the program found.
print('Image shape (height, width, colour channels):', image.shape)
```

## 2. Look closely
The first two numbers tell us the image size. The final **3** means red, green, and blue channels.


```python
# Store this result in a named variable so later lines can use it.
y, x = 180, 250
# Print a helpful result so you can inspect what the program found.
print(f'Pixel at (x={x}, y={y}) has RGB values:', image[y, x])

# Create a plotting area where the image results will be displayed.
plt.figure(figsize=(5, 5))
# Display this image or result in the current plotting area.
plt.imshow(image)
# Draw a visible marker at the selected pixel position.
plt.scatter([x], [y], color='yellow', s=90)
# Add a title so the picture or result is easy to identify.
plt.title('One pixel, marked in yellow')
# Hide the graph axes because they are not useful for this picture.
plt.axis('off');
```

## 3. Make simple edits


```python
# Store this result in a named variable so later lines can use it.
gray = np.mean(image, axis=2)
# Store this result in a named variable so later lines can use it.
crop = image[80:360, 130:410]
# Store this result in a named variable so later lines can use it.
small = image[::3, ::3]
# Store this result in a named variable so later lines can use it.
flipped = image[:, ::-1]

# Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
# Repeat the indented instructions once for every item in this group.
for ax, picture, title in zip(axes, [gray, crop, small, flipped], ['Grayscale', 'Crop', 'Smaller', 'Flipped']):
    # Display this image or result in the current plotting area.
    ax.imshow(picture, cmap='gray' if picture.ndim == 2 else None)
    # Add a title so the picture or result is easy to identify.
    ax.set_title(title)
    # Hide the graph axes because they are not useful for this picture.
    ax.axis('off')
# Show all of the plots you have prepared.
plt.show()
```
