## 1. Load a picture
An image is a grid of tiny colour squares called **pixels**. A colour pixel usually stores three numbers: red, green, and blue.


```python
# BEGINNER NOTE: Import numpy so this notebook can use its ready-made tools.
import numpy as np
# BEGINNER NOTE: Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# BEGINNER NOTE: Import the specific tools this lesson needs from skimage.
from skimage import data

# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
image = data.astronaut()  # an RGB image included with scikit-image
# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
plt.figure(figsize=(6, 6))
# BEGINNER NOTE: Display this image or result in the current plotting area.
plt.imshow(image)
# BEGINNER NOTE: Hide the graph axes because they are not useful for this picture.
plt.axis('off')
# BEGINNER NOTE: Show all of the plots you have prepared.
plt.show()

# BEGINNER NOTE: Print a helpful result so you can inspect what the program found.
print('Image shape (height, width, colour channels):', image.shape)
```

## 2. Look closely
The first two numbers tell us the image size. The final **3** means red, green, and blue channels.


```python
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
y, x = 180, 250
# BEGINNER NOTE: Print a helpful result so you can inspect what the program found.
print(f'Pixel at (x={x}, y={y}) has RGB values:', image[y, x])

# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
plt.figure(figsize=(5, 5))
# BEGINNER NOTE: Display this image or result in the current plotting area.
plt.imshow(image)
# BEGINNER NOTE: Draw a visible marker at the selected pixel position.
plt.scatter([x], [y], color='yellow', s=90)
# BEGINNER NOTE: Add a title so the picture or result is easy to identify.
plt.title('One pixel, marked in yellow')
# BEGINNER NOTE: Hide the graph axes because they are not useful for this picture.
plt.axis('off');
```

## 3. Make simple edits


```python
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
gray = np.mean(image, axis=2)
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
crop = image[80:360, 130:410]
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
small = image[::3, ::3]
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
flipped = image[:, ::-1]

# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
# BEGINNER NOTE: Repeat the indented instructions once for every item in this group.
for ax, picture, title in zip(axes, [gray, crop, small, flipped], ['Grayscale', 'Crop', 'Smaller', 'Flipped']):
    # BEGINNER NOTE: Display this image or result in the current plotting area.
    ax.imshow(picture, cmap='gray' if picture.ndim == 2 else None)
    # BEGINNER NOTE: Add a title so the picture or result is easy to identify.
    ax.set_title(title)
    # BEGINNER NOTE: Hide the graph axes because they are not useful for this picture.
    ax.axis('off')
# BEGINNER NOTE: Show all of the plots you have prepared.
plt.show()
```
