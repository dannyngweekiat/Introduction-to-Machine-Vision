# Lesson 7: Handwritten Digit Recognition

## Your mission
Train a model to recognise handwritten digits and study its mistakes.

**Key words:** digit · train/test split · confusion matrix

Run each code cell in order. Cells marked **Try it yourself** are yours to change.


```python
# BEGINNER NOTE: Import numpy so this notebook can use its ready-made tools.
import numpy as np
# BEGINNER NOTE: Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# BEGINNER NOTE: Import the specific tools this lesson needs from sklearn.datasets.
from sklearn.datasets import load_digits
# BEGINNER NOTE: Import the specific tools this lesson needs from sklearn.model_selection.
from sklearn.model_selection import train_test_split
# BEGINNER NOTE: Import the specific tools this lesson needs from sklearn.neighbors.
from sklearn.neighbors import KNeighborsClassifier
# BEGINNER NOTE: Import the specific tools this lesson needs from sklearn.metrics.
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay

# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
digits = load_digits()
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
X, y = digits.data, digits.target

# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(2, 6, figsize=(11, 4))
# BEGINNER NOTE: Repeat the indented instructions once for every item in this group.
for ax, picture, label in zip(axes.flat, digits.images[:12], y[:12]):
    # BEGINNER NOTE: Display this image or result in the current plotting area.
    ax.imshow(picture, cmap='gray_r')
    # BEGINNER NOTE: Add a title so the picture or result is easy to identify.
    ax.set_title(f'Digit {label}')
    # BEGINNER NOTE: Hide the graph axes because they are not useful for this picture.
    ax.axis('off')
```

## 1. Keep some images secret
We train on one group and test on another. A fair test uses images the model did not practise on.


```python
# BEGINNER NOTE: Split the examples so some are saved for a fair test later.
X_train, X_test, y_train, y_test = train_test_split(
    # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
    X, y, test_size=0.25, random_state=12, stratify=y
# BEGINNER NOTE: Finish this command and save or use its completed result.
)

# BEGINNER NOTE: Create an AI model that compares each example with nearby examples.
model = KNeighborsClassifier(n_neighbors=3)
# BEGINNER NOTE: Let the AI model learn patterns from the training examples.
model.fit(X_train, y_train)
# BEGINNER NOTE: Ask the trained AI model to make predictions for these examples.
predicted = model.predict(X_test)
# BEGINNER NOTE: Print a helpful result so you can inspect what the program found.
print('Test accuracy:', f'{accuracy_score(y_test, predicted):.1%}')
```

## 2. See correct and incorrect predictions


```python
# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(2, 6, figsize=(11, 4))
# BEGINNER NOTE: Repeat the indented instructions once for every item in this group.
for ax, picture, truth, guess in zip(axes.flat, X_test[:12].reshape(-1, 8, 8), y_test[:12], predicted[:12]):
    # BEGINNER NOTE: Display this image or result in the current plotting area.
    ax.imshow(picture, cmap='gray_r')
    # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
    colour = 'green' if truth == guess else 'crimson'
    # BEGINNER NOTE: Add a title so the picture or result is easy to identify.
    ax.set_title(f'True {truth} | AI {guess}', color=colour, fontsize=10)
    # BEGINNER NOTE: Hide the graph axes because they are not useful for this picture.
    ax.axis('off')
# BEGINNER NOTE: Show all of the plots you have prepared.
plt.show()
```

## 3. Find confusing digit pairs
A confusion matrix shows what the true digit was (row) and what the AI guessed (column).


```python
# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
fig, ax = plt.subplots(figsize=(7, 6))
# BEGINNER NOTE: Draw a table that shows which digit predictions get mixed up.
ConfusionMatrixDisplay.from_predictions(y_test, predicted, cmap='Blues', colorbar=False, ax=ax)
# BEGINNER NOTE: Add a title so the picture or result is easy to identify.
ax.set_title('Which digits get mixed up?')
# BEGINNER NOTE: Show all of the plots you have prepared.
plt.show()
```
