```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

coins = data.coins()
plt.figure(figsize=(7, 5))
plt.imshow(coins, cmap='gray')
plt.title('Can a computer count these coins?')
plt.axis('off');
```

## 1. Separate objects from the background
A **threshold** turns pixels into black or white. Small blur helps reduce speckled noise.


```python
smooth = cv2.GaussianBlur(coins, (5, 5), 0)
_, binary = cv2.threshold(smooth, 105, 255, cv2.THRESH_BINARY_INV)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(coins, cmap='gray'); axes[0].set_title('Original')
axes[1].imshow(binary, cmap='gray'); axes[1].set_title('Threshold image')
for ax in axes: ax.axis('off')
```

## 2. Find contours and draw boxes
A **contour** is an outline around connected pixels. We filter very tiny outlines because they are usually noise.


```python
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
result = cv2.cvtColor(coins, cv2.COLOR_GRAY2RGB)
count = 0

for contour in contours:
    area = cv2.contourArea(contour)
    if area > 90:  # remove tiny noise blobs
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x, y), (x + w, y + h), (255, 80, 0), 2)
        count += 1

plt.figure(figsize=(8, 6))
plt.imshow(result)
plt.title(f'Objects counted: {count}')
plt.axis('off');
```
