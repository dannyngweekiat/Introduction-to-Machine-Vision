## Your mission
Use simple rules to find and count objects in an image.

**Key words:** threshold · contour · bounding box

Run each code cell in order. Cells marked **Try it yourself** are yours to change.


```python
# BEGINNER NOTE: Import cv2 so this notebook can use its ready-made tools.
import cv2
# BEGINNER NOTE: Import numpy so this notebook can use its ready-made tools.
import numpy as np
# BEGINNER NOTE: Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# BEGINNER NOTE: Import the specific tools this lesson needs from skimage.
from skimage import data

# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
coins = data.coins()
# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
plt.figure(figsize=(7, 5))
# BEGINNER NOTE: Display this image or result in the current plotting area.
plt.imshow(coins, cmap='gray')
# BEGINNER NOTE: Add a title so the picture or result is easy to identify.
plt.title('Can a computer count these coins?')
# BEGINNER NOTE: Hide the graph axes because they are not useful for this picture.
plt.axis('off');
```

## 1. Separate objects from the background
A **threshold** turns pixels into black or white. Small blur helps reduce speckled noise.


```python
# BEGINNER NOTE: Blur nearby pixels together to reduce small details and noise.
smooth = cv2.GaussianBlur(coins, (5, 5), 0)
# BEGINNER NOTE: Turn the picture into a black-and-white image using this brightness cutoff.
_, binary = cv2.threshold(smooth, 105, 255, cv2.THRESH_BINARY_INV)

# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# BEGINNER NOTE: Display this image or result in the current plotting area.
axes[0].imshow(coins, cmap='gray'); axes[0].set_title('Original')
# BEGINNER NOTE: Display this image or result in the current plotting area.
axes[1].imshow(binary, cmap='gray'); axes[1].set_title('Threshold image')
# BEGINNER NOTE: Repeat the indented instructions once for every item in this group.
for ax in axes: ax.axis('off')
```

## 2. Find contours and draw boxes
A **contour** is an outline around connected pixels. We filter very tiny outlines because they are usually noise.


```python
# BEGINNER NOTE: Find the outlines of the separate shapes in the mask.
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# BEGINNER NOTE: Convert the image into a colour format that suits the next task.
result = cv2.cvtColor(coins, cv2.COLOR_GRAY2RGB)
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
count = 0

# BEGINNER NOTE: Repeat the indented instructions once for every item in this group.
for contour in contours:
    # BEGINNER NOTE: Measure how many pixels are inside this shape's outline.
    area = cv2.contourArea(contour)
    # BEGINNER NOTE: Check this condition and run the indented code only when it is true.
    if area > 90:  # remove tiny noise blobs
        # BEGINNER NOTE: Find the smallest rectangle that can surround this shape.
        x, y, w, h = cv2.boundingRect(contour)
        # BEGINNER NOTE: Draw a rectangle around the shape that was found.
        cv2.rectangle(result, (x, y), (x + w, y + h), (255, 80, 0), 2)
        # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
        count += 1

# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
plt.figure(figsize=(8, 6))
# BEGINNER NOTE: Display this image or result in the current plotting area.
plt.imshow(result)
# BEGINNER NOTE: Add a title so the picture or result is easy to identify.
plt.title(f'Objects counted: {count}')
# BEGINNER NOTE: Hide the graph axes because they are not useful for this picture.
plt.axis('off');
```
