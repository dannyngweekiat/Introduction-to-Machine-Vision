## 1. Make labelled examples
AI does not “see” like a person. It learns patterns from examples with answers called **labels**.


```python
# BEGINNER NOTE: Import numpy so this notebook can use its ready-made tools.
import numpy as np
# BEGINNER NOTE: Import matplotlib.pyplot so this notebook can use its ready-made tools.
import matplotlib.pyplot as plt
# BEGINNER NOTE: Import the specific tools this lesson needs from sklearn.model_selection.
from sklearn.model_selection import train_test_split
# BEGINNER NOTE: Import the specific tools this lesson needs from sklearn.neighbors.
from sklearn.neighbors import KNeighborsClassifier
# BEGINNER NOTE: Import the specific tools this lesson needs from sklearn.metrics.
from sklearn.metrics import accuracy_score

# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
rng = np.random.default_rng(7)

# BEGINNER NOTE: Define a reusable set of instructions with a clear name.
def make_example(label):
    # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
    image = np.ones((32, 32, 3), dtype=np.uint8) * 245
    # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
    yy, xx = np.ogrid[:32, :32]
    # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
    cx, cy = rng.integers(11, 21, size=2)
    # BEGINNER NOTE: Check this condition and run the indented code only when it is true.
    if label == 'circle':
        # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
        mask = (xx - cx) ** 2 + (yy - cy) ** 2 < rng.integers(45, 80)
        # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
        image[mask] = [220, 55, 65]
    # BEGINNER NOTE: Run this alternative when the earlier condition was not true.
    else:
        # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
        size = rng.integers(9, 14)
        # BEGINNER NOTE: Store this result in a named variable so later lines can use it.
        image[cy-size//2:cy+size//2, cx-size//2:cx+size//2] = [50, 110, 220]
    # BEGINNER NOTE: Send the finished value back to the code that called this function.
    return image

# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
images = [make_example('circle') for _ in range(40)] + [make_example('square') for _ in range(40)]
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
labels = np.array(['circle'] * 40 + ['square'] * 40)

# BEGINNER NOTE: Create a plotting area where the image results will be displayed.
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
# BEGINNER NOTE: Repeat the indented instructions once for every item in this group.
for ax, picture, label in zip(axes.flat, images[:5] + images[40:45], list(labels[:5]) + list(labels[40:45])):
    # BEGINNER NOTE: Display this image or result in the current plotting area.
    ax.imshow(picture); ax.set_title(label); ax.axis('off')
```

## 2. Turn pictures into numbers and train
We use the average red, green, and blue values as simple **features**. KNN predicts the label of nearby examples.


```python
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
features = np.array([picture.mean(axis=(0, 1)) for picture in images])
# BEGINNER NOTE: Split the examples so some are saved for a fair test later.
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.25, random_state=3, stratify=labels)

# BEGINNER NOTE: Create an AI model that compares each example with nearby examples.
model = KNeighborsClassifier(n_neighbors=3)
# BEGINNER NOTE: Let the AI model learn patterns from the training examples.
model.fit(X_train, y_train)
# BEGINNER NOTE: Ask the trained AI model to make predictions for these examples.
predictions = model.predict(X_test)
# BEGINNER NOTE: Print a helpful result so you can inspect what the program found.
print('Test accuracy:', round(accuracy_score(y_test, predictions) * 100, 1), '%')
# BEGINNER NOTE: Print a helpful result so you can inspect what the program found.
print('First five predictions:', predictions[:5])
```

## 3. Find a mistake
Accuracy is useful, but it does not tell the whole story. We should always inspect examples the model gets wrong.


```python
# Make a tricky purple circle. It may confuse a model that only learned red circles.
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
tricky = np.ones((32, 32, 3), dtype=np.uint8) * 245
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
yy, xx = np.ogrid[:32, :32]
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
tricky[(xx - 16) ** 2 + (yy - 16) ** 2 < 70] = [130, 70, 180]
# BEGINNER NOTE: Store this result in a named variable so later lines can use it.
tricky_feature = tricky.mean(axis=(0, 1)).reshape(1, -1)

# BEGINNER NOTE: Display this image or result in the current plotting area.
plt.imshow(tricky); plt.axis('off'); plt.show()
# BEGINNER NOTE: Print a helpful result so you can inspect what the program found.
print('Prediction for the purple circle:', model.predict(tricky_feature)[0])
```
