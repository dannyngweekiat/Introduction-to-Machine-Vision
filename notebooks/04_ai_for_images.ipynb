## 1. Make labelled examples
AI does not “see” like a person. It learns patterns from examples with answers called **labels**.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

rng = np.random.default_rng(7)

def make_example(label):
    image = np.ones((32, 32, 3), dtype=np.uint8) * 245
    yy, xx = np.ogrid[:32, :32]
    cx, cy = rng.integers(11, 21, size=2)
    if label == 'circle':
        mask = (xx - cx) ** 2 + (yy - cy) ** 2 < rng.integers(45, 80)
        image[mask] = [220, 55, 65]
    else:
        size = rng.integers(9, 14)
        image[cy-size//2:cy+size//2, cx-size//2:cx+size//2] = [50, 110, 220]
    return image

images = [make_example('circle') for _ in range(40)] + [make_example('square') for _ in range(40)]
labels = np.array(['circle'] * 40 + ['square'] * 40)

fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for ax, picture, label in zip(axes.flat, images[:5] + images[40:45], list(labels[:5]) + list(labels[40:45])):
    ax.imshow(picture); ax.set_title(label); ax.axis('off')
```

## 2. Turn pictures into numbers and train
We use the average red, green, and blue values as simple **features**. KNN predicts the label of nearby examples.


```python
features = np.array([picture.mean(axis=(0, 1)) for picture in images])
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.25, random_state=3, stratify=labels)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print('Test accuracy:', round(accuracy_score(y_test, predictions) * 100, 1), '%')
print('First five predictions:', predictions[:5])
```

## 3. Find a mistake
Accuracy is useful, but it does not tell the whole story. We should always inspect examples the model gets wrong.


```python
# Make a tricky purple circle. It may confuse a model that only learned red circles.
tricky = np.ones((32, 32, 3), dtype=np.uint8) * 245
yy, xx = np.ogrid[:32, :32]
tricky[(xx - 16) ** 2 + (yy - 16) ** 2 < 70] = [130, 70, 180]
tricky_feature = tricky.mean(axis=(0, 1)).reshape(1, -1)

plt.imshow(tricky); plt.axis('off'); plt.show()
print('Prediction for the purple circle:', model.predict(tricky_feature)[0])
```
